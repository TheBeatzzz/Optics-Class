```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Generate an array of 1000 time points from 0 to 10 fs
time_fs = np.linspace(0, 10, 1000)

# Display the first few values
print(time_fs[:10])
```

```python
# Wavelength in nanometers (nm)
wavelength_nm = 500

# Convert wavelength to meters (m)
wavelength_m = wavelength_nm * 1e-9

# Speed of light in meters per second (m/s)
c = 299792458

# Frequency of the wave (Hz)
frequency_hz = c / wavelength_m

# Angular frequency (rad/s)
angular_frequency_rad_s = 2 * np.pi * frequency_hz

# Amplitude of the electric field (V/m)
amplitude_v_m = 1  # Using 1V amplitude as requested

# Electric field as a function of time (assuming propagation in one direction)
# We will assume the spatial part is constant at x=0 for this plot
electric_field_t = amplitude_v_m * np.sin(angular_frequency_rad_s * time_fs * 1e-15) # time_fs is in fs, so convert to seconds

# Plot the electric field as a function of time
plt.figure(figsize=(10, 6))
plt.plot(time_fs, electric_field_t)
plt.xlabel("Time (fs)")
plt.ylabel("Electric Field Amplitude (V/m)")
plt.title("Electric Field of a 500 nm Wavelength Wave")
plt.grid(True)
plt.show()
```

```python
# Define the spatial window in micrometers (µm)
spatial_window_um = 3

# Generate an array of spatial points from 0 to 10 µm (convert to meters for calculations)
spatial_m = np.linspace(0, spatial_window_um * 1e-6, 1000) # 1000 points for smooth curve

# Wavelength in meters (already calculated in previous cell, but redefined for clarity)
wavelength_m = 500 * 1e-9

# Wavenumber (k) in rad/m
wavenumber_rad_m = 2 * np.pi / wavelength_m

# Amplitude of the electric field (V/m) (already defined)
amplitude_v_m = 1

# Electric field as a function of space (at a fixed time, e.g., t=0)
electric_field_x = amplitude_v_m * np.sin(wavenumber_rad_m * spatial_m)

# Plot the electric field as a function of space
plt.figure(figsize=(10, 6))
plt.plot(spatial_m * 1e6, electric_field_x) # Convert spatial_m back to µm for plotting
plt.xlabel("Space (µm)")
plt.ylabel("Electric Field Amplitude (V/m)")
plt.title("Electric Field of a 500 nm Wavelength Wave in Space")
plt.grid(True)
plt.show()
```
